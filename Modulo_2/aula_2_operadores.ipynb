Operadores Aritiméticos
--
- Os operadores possuem comportamentos diferentes dependendo do tipo de variáveis

```python
# Adição númerica
print(a+b)
```

```python
# Adição de strings
print(c)
print(d)
print(c+d)
```

```python
# Subtração
print(a-b)
```

```python
# Divisão
print(b/a)
```

```python
# Parte inteira da divisão
print(b//a)
```

```python
# Resto da Divisão
print(b%a)
```

```python
# Multiplicação
print(a*b)
```

```python
# Potência
print(a**3)
```

Operadores Relacionais
--

```python
# Teste de igualdade númerica
a == b
```

```python
# Teste de diferença númerica
a != b
```

```python
# Teste de igualdade de strings
c == d, "eu" == "eu", "eu" == "Eu"
```

```python
# Teste de diferença de string
a != b
```

```python
# Maior
a > b
```

```python
# Menor
a < b 
```

```python
# Maior Igual
a >= b
```

```python
# Menor Igual
a <= b
```
