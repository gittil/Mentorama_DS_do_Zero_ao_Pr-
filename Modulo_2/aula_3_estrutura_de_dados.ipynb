#Conjuntos

- Conjuntos não possuem noção de ordem, por isso seus elementos não podem ser a acessados por colchetes [] nem por slices
- Conjuntos só aceitam tipos de dados imutáveis como inteiros, floats, tuplas e strings, não aceitando listas, dicionários e outros conjuntos(set)

**Criando conjuntos**

```python
#Usa chaves também, mas não temos pares e sim elementos
primeiro_conjunto = {2,4,6,8}
```

```python
#Não mantém a ordem, conjunto não é um objeto ordenado
print(primeiro_conjunto)
```

```python
type(primeiro_conjunto)
```

```python
len(primeiro_conjunto)
```

```python
#função set()
#conjuntos não permitem elementos repetidos

lista_ = [1,2,5,6,6,7,8,8,]

print(len(lista_))

print(set(lista_))

print(len(set(lista_)))
```

```python
#podemos criar conjuntos vazios

conj_vazio = set() #é dessa forma que se cria um conjunto vazio
dict_vazio = {}
```

```python
type(conj_vazio)
```

```python
type(dict_vazio)
```

**Operações entre conjuntos**
- Suponha que vc tem duas listas e quer saber que elementos estão nas duas, ista é, estão na primeira lista e na segunda lista?

```python
pares = [2,4,6,8,10]
multiplos_4 = [4,8,12,16]

pares = set(pares)
multiplos_4 = set(multiplos_4)
```

```python
print(type(pares))
print(type(multiplos_4))
```

```python
#Intersecção: Elementos que estão no conjuntos PARES e no conjunto MULTIPLOS_4

pares.intersection(multiplos_4)
```

```python
#União: faz a união das duas listas, mas sem repetir valores

pares.union(multiplos_4)
```

```python
#Diferença: Elementos que estão em PARES mas não estão em MULTIPLOS_4

pares.difference(multiplos_4)
```
